## Draft EU Artificial Intelligence Act revision

The goal of this exercise is to get an overview on the Draft EU Artificial Intelligence Act (Draft EU AI Act) and to critically examine some of its provisions. In order to be able to answer the following questions, you should read M. Veale/F. Zuiderveen Borgesius, Demystifying the Draft EU Artificial Intelligence Act, Computer Law Review International (CRi) 22 (2021), 97-112 (you may leave out paragraphs 49-92). 
In addition, although not strictly required, it is recommended to take a look at the Draft EU Artificial Intelligence Act itself: Proposal for a REGULATION OF THE EUROPEAN PARLIAMENT AND OF THE COUNCIL LAYING DOWN HARMONISED RULES ON ARTIFICIAL INTELLIGENCE (ARTIFICIAL INTELLIGENCE ACT) AND AMENDING CERTAIN UNION LEGISLATIVE ACTS COM/2021/206 final (Link: https://eur-lex.europa.eu/legal-content/EN/TXT/?uri=CELEX%3A52021PC0206) (relevant pages: 38-84)
